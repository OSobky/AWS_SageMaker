```python
df = df.dropna()
```

```python
df['Training Hours Start'].unique()
```

```python
df.info()
```

```python
df["Coach"].replace({"mostafa": "Mostafa", "nour": "Nour", "ibrahim": "Ibrahim"}, inplace=True)
```

```python
df['Trainig Days'].unique()
```

```python
for train_days in df['Trainig Days'].unique():
    output_days = pd.DataFrame()
    df_days = df[df['Trainig Days'] == train_days]
    output_days['Coach'] =  df_days['Coach'].unique()
    
    for time in df_days['Training Hours Start'].unique():
        output_days[time] = ""
    
    for Coach in output_days['Coach'].unique():
        for time in df_days['Training Hours Start'].unique():
            list_clients_time_coach = df_days[(df_days['Coach'] == Coach) & (df_days['Training Hours Start'] == time)]['Client Name'].values.tolist()
            names = ""
            for name in list_clients_time_coach:
                names +=  name + " -- "
            print(names)
            output_days.loc[output_days["Coach"] == Coach, [time]] = output_days.loc[output_days["Coach"] == Coach, [time]] + names
    output_days.to_csv(train_days + "output.csv")
    
    
```
